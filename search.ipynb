# Fetching articles from Le Monde and Le Figaro for 2 query

⚠️ You need to run a selenium docker on port 4444 ⚠️
> docker run -d -p 4444:4444 --shm-size="2g" selenium/standalone-chrome:4.15.0-20231122

```python
import lemonde
import figaro
import pandas as pd
import os 

COOKIE_LM = "/mnt/2To/jupyter_data/socc/www.lemonde.fr_cookies.txt"
COOKIE_FG = "/mnt/2To/jupyter_data/socc/www.lefigaro.fr_cookies.txt"

#### SEARCH ONE ####
search_1 = "Serge D."
date_from_1 = "01/03/2023"
date_to_1 = "01/05/2023"
#### SEARCH TWO ####
search_2 = "Rémi Fraisse"
date_from_2 = "01/10/2014"
date_to_2 = "01/12/2014"

df_lm_1 = lemonde.fetch_lemonde(search_1, date_from_1, date_to_1) #25.03.2023
df_lm_2 = lemonde.fetch_lemonde(search_2, date_from_2, date_to_2) #25.10.2014

df_fg_1 = figaro.fetch_figaro(search_1, date_from_1, date_to_1) #25.03.2023
df_fg_2 = figaro.fetch_figaro(search_2, date_from_2, date_to_2) #25.10.2014
#Do some filtering
df_fg_1 = df_fg_1[df_fg_1['theme']!="Prénoms"] 
df_fg_2 = df_fg_2[df_fg_2['theme']!="Prénoms"]
df_fg_1['og_website'] = df_fg_1['url'].str.contains('www.lefigaro.fr')
df_fg_2['og_website'] = df_fg_2['url'].str.contains('www.lefigaro.fr')

#Display the data
display(df_lm_1)
display(df_lm_2)
display(df_fg_1)
display(df_fg_2)
```

```python
#Get article content Le Monde
df_lm_1_content = lemonde.article_contents(df_lm_1[df_lm_1['url'].str.contains('https://www.lemonde.fr/podcasts')==False]['url'].tolist(), COOKIE_LM)
df_lm_2_content = lemonde.article_contents(df_lm_2[df_lm_2['url'].str.contains('https://www.lemonde.fr/podcasts')==False]['url'].tolist(), COOKIE_LM)
#Get article content Le Figaro
df_fg_1_content = figaro.article_contents(df_fg_1[df_fg_1['og_website']==True]['url'].tolist(), COOKIE_FG)
df_fg_2_content = figaro.article_contents(df_fg_2[df_fg_2['og_website']==True]['url'].tolist(), COOKIE_FG)
```

```python
#Merge the data
df_lm_1 = pd.merge(df_lm_1, pd.DataFrame(df_lm_1_content), on='url', how='left')
df_lm_2 = pd.merge(df_lm_2, pd.DataFrame(df_lm_2_content), on='url', how='left')
df_fg_1 = pd.merge(df_fg_1, pd.DataFrame(df_fg_1_content), on='url', how='left')
df_fg_2 = pd.merge(df_fg_2, pd.DataFrame(df_fg_2_content), on='url', how='left')
```

```python
#Save the data
FOLDER = '/mnt/2To/jupyter_data/socc/'
os.mkdir(os.path.join(FOLDER, search_1.replace(" ","_").replace(".","")+"-"+search_2.replace(" ","_").replace(".","")))
path_fg_1 = os.path.join(FOLDER, search_1.replace(" ","_").replace(".","")+"-"+search_2.replace(" ","_").replace(".",""), "figaro_"+search_1.replace(" ","_")+"_"+date_from_1.replace("/","_")+"_"+date_to_1.replace("/","_")+".csv")
path_fg_2 = os.path.join(FOLDER, search_1.replace(" ","_").replace(".","")+"-"+search_2.replace(" ","_").replace(".",""), "figaro_"+search_2.replace(" ","_")+"_"+date_from_2.replace("/","_")+"_"+date_to_2.replace("/","_")+".csv")
path_lm_1 = os.path.join(FOLDER, search_1.replace(" ","_").replace(".","")+"-"+search_2.replace(" ","_").replace(".",""), "lemonde_"+search_1.replace(" ","_")+"_"+date_from_1.replace("/","_")+"_"+date_to_1.replace("/","_")+".csv")
path_lm_2 = os.path.join(FOLDER, search_1.replace(" ","_").replace(".","")+"-"+search_2.replace(" ","_").replace(".",""), "lemonde_"+search_2.replace(" ","_")+"_"+date_from_2.replace("/","_")+"_"+date_to_2.replace("/","_")+".csv")
df_fg_1.to_csv(path_fg_1, index=False)
df_fg_2.to_csv(path_fg_2, index=False)
df_lm_1.to_csv(path_lm_1, index=False)
df_lm_2.to_csv(path_lm_2, index=False)
df_fg_1.to_pickle(path_fg_1.replace(".csv",".pkl"))
df_fg_2.to_pickle(path_fg_2.replace(".csv",".pkl"))
df_lm_1.to_pickle(path_lm_1.replace(".csv",".pkl"))
df_lm_2.to_pickle(path_lm_2.replace(".csv",".pkl"))
```
